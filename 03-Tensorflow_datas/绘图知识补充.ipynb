## Matplotlib常用图像类型

1. **折线图 (Line Plot)**  
  用于显示数据的趋势或变化，通常用于连续数据的可视化。

2. **散点图 (Scatter Plot)**  
  用于显示两个变量之间的关系，适合分析数据的分布和相关性。

3. **柱状图 (Bar Chart)**  
  用于比较不同类别的数据大小，适合离散数据的可视化。

4. **直方图 (Histogram)**  
  用于显示数据的分布情况，特别是频率分布。

5. **饼图 (Pie Chart)**  
  用于显示各部分占总体的比例，适合展示百分比数据。

6. **箱线图 (Box Plot)**  
  用于显示数据的分布特征，包括中位数、四分位数和异常值。

7. **热图 (Heatmap)**  
  用于显示矩阵数据的值大小，通过颜色深浅来表示。

8. **误差条图 (Error Bar Plot)**  
  用于显示数据的不确定性或误差范围。

9. **面积图 (Area Chart)**  
  类似折线图，但通过填充线下区域来表示数据的累积值。

10. **极坐标图 (Polar Plot)**  
   用于绘制极坐标系中的数据，适合周期性数据的可视化。

11. **3D图 (3D Plot)**  
   用于显示三维数据的关系，包括3D散点图、3D曲面图等。

这些图像类型可以根据数据特点和分析需求选择使用，帮助更直观地理解数据。

```python
import numpy as np

import matplotlib.pyplot as plt

# 设置中文字体支持（需要安装 SimHei 字体）
# plt.rcParams['font.sans-serif'] = ['SimHei']  # 用黑体显示中文
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']  # 用黑体显示中文
# plt.rcParams['font.sans-serif'] = ['STHeiti']  # 用黑体显示中文
plt.rcParams['axes.unicode_minus'] = False   # 正常显示负号

# 生成数据
x = np.linspace(0, 10, 100)  # 在 0 到 10 之间生成 100 个等间距点
y1 = np.sin(x)               # y1 为正弦函数
y2 = np.cos(x)               # y2 为余弦函数

# 创建图形和子图
plt.figure(figsize=(10, 6))  # 设置图形大小为 10x6 英寸

# 绘制正弦曲线
plt.plot(x, y1, label="正弦函数", color="blue", linestyle="-", linewidth=2)

# 绘制余弦曲线
plt.plot(x, y2, label="余弦函数", color="red", linestyle="--", linewidth=2)

# 添加标题和轴标签
plt.title("正弦函数与余弦函数的对比", fontsize=16)  # 图表标题
plt.xlabel("自变量 x", fontsize=14)              # x 轴标签
plt.ylabel("函数值", fontsize=14)                # y 轴标签

# 添加网格线
plt.grid(True, linestyle=":", color="gray", alpha=0.7)

# 添加图例
plt.legend(fontsize=12, loc="upper right")  # 图例位置为右上角

# 设置坐标轴范围
plt.xlim(0, 10)  # x 轴范围
plt.ylim(-1.5, 1.5)  # y 轴范围

# 保存图表为高分辨率图片
plt.savefig("scientific_plot.png", dpi=300, bbox_inches="tight")

# 显示图表
plt.show()
```

```python
# Line Plot (折线图)
plt.figure(figsize=(8, 5))  # 设置图形大小
plt.plot(x, y1, label="Sine Function", color="blue")  # 绘制正弦函数折线图
plt.title("Line Plot - Sine Function", fontsize=16)  # 设置标题
plt.xlabel("Independent Variable x", fontsize=14)  # 设置 x 轴标签
plt.ylabel("Function Value", fontsize=14)  # 设置 y 轴标签
plt.legend(fontsize=12)  # 添加图例
plt.grid(True, linestyle="--", alpha=0.7)  # 添加网格线
plt.show()  # 显示图形

# Scatter Plot (散点图)
plt.figure(figsize=(8, 5))  # 设置图形大小
plt.scatter(x, y1, label="Sine Function", color="green", s=10)  # 绘制正弦函数散点图
plt.title("Scatter Plot - Sine Function", fontsize=16)  # 设置标题
plt.xlabel("Independent Variable x", fontsize=14)  # 设置 x 轴标签
plt.ylabel("Function Value", fontsize=14)  # 设置 y 轴标签
plt.legend(fontsize=12)  # 添加图例
plt.grid(True, linestyle="--", alpha=0.7)  # 添加网格线
plt.show()  # 显示图形

# Bar Chart (柱状图)
plt.figure(figsize=(8, 5))  # 设置图形大小
plt.bar(x[::10], y1[::10], label="Sine Function", color="orange", width=0.3)  # 绘制正弦函数柱状图（采样点）
plt.title("Bar Chart - Sine Function (Sample Points)", fontsize=16)  # 设置标题
plt.xlabel("Independent Variable x", fontsize=14)  # 设置 x 轴标签
plt.ylabel("Function Value", fontsize=14)  # 设置 y 轴标签
plt.legend(fontsize=12)  # 添加图例
plt.grid(True, linestyle="--", alpha=0.7)  # 添加网格线
plt.show()  # 显示图形

# Histogram (直方图)
plt.figure(figsize=(8, 5))  # 设置图形大小
plt.hist(y1, bins=20, color="purple", alpha=0.7, label="Sine Function Value Distribution")  # 绘制正弦函数值分布直方图
plt.title("Histogram - Sine Function Value Distribution", fontsize=16)  # 设置标题
plt.xlabel("Function Value", fontsize=14)  # 设置 x 轴标签
plt.ylabel("Frequency", fontsize=14)  # 设置 y 轴标签
plt.legend(fontsize=12)  # 添加图例
plt.grid(True, linestyle="--", alpha=0.7)  # 添加网格线
plt.show()  # 显示图形

# Box Plot (箱线图)
plt.figure(figsize=(8, 5))  # 设置图形大小
plt.boxplot(y1, vert=False, patch_artist=True, boxprops=dict(facecolor="cyan"))  # 绘制正弦函数值分布箱线图
plt.title("Box Plot - Sine Function Value Distribution", fontsize=16)  # 设置标题
plt.xlabel("Function Value", fontsize=14)  # 设置 x 轴标签
plt.grid(True, linestyle="--", alpha=0.7)  # 添加网格线
plt.show()  # 显示图形
```

```python
# 定义线性函数 y = ax + b
a = 0.1  # 斜率
b = 0.5  # 截距
y_linear = a * x + b  # 计算线性函数值

# 创建一个包含四个子图的图形
fig, axes = plt.subplots(2, 2, figsize=(12, 10))  # 2x2 布局，图形大小为 12x10 英寸

# 折线图 (Line Plot)
axes[0, 0].plot(x, y1, label="Sine Function", color="blue", linestyle="-")  # 绘制正弦函数
axes[0, 0].plot(x, y_linear, label="Linear Function", color="red", linestyle="--")  # 绘制线性函数
axes[0, 0].set_title("Line Plot", fontsize=14)  # 设置标题
axes[0, 0].set_xlabel("x", fontsize=12)  # 设置 x 轴标签
axes[0, 0].set_ylabel("y", fontsize=12)  # 设置 y 轴标签
axes[0, 0].legend(fontsize=10)  # 添加图例
axes[0, 0].grid(True, linestyle="--", alpha=0.7)  # 添加网格线

# 散点图 (Scatter Plot)
axes[0, 1].scatter(x, y1, label="Sine Function", color="green", s=10)  # 绘制正弦函数散点图
axes[0, 1].scatter(x, y_linear, label="Linear Function", color="orange", s=10)  # 绘制线性函数散点图
axes[0, 1].set_title("Scatter Plot", fontsize=14)  # 设置标题
axes[0, 1].set_xlabel("x", fontsize=12)  # 设置 x 轴标签
axes[0, 1].set_ylabel("y", fontsize=12)  # 设置 y 轴标签
axes[0, 1].legend(fontsize=10)  # 添加图例
axes[0, 1].grid(True, linestyle="--", alpha=0.7)  # 添加网格线

# 柱状图 (Bar Chart)
axes[1, 0].bar(x[::10], y1[::10], label="Sine Function", color="blue", width=0.3)  # 绘制正弦函数柱状图（采样点）
axes[1, 0].bar(x[::10] + 0.3, y_linear[::10], label="Linear Function", color="red", width=0.3)  # 绘制线性函数柱状图（采样点）
axes[1, 0].set_title("Bar Chart", fontsize=14)  # 设置标题
axes[1, 0].set_xlabel("x", fontsize=12)  # 设置 x 轴标签
axes[1, 0].set_ylabel("y", fontsize=12)  # 设置 y 轴标签
axes[1, 0].legend(fontsize=10)  # 添加图例
axes[1, 0].grid(True, linestyle="--", alpha=0.7)  # 添加网格线

# 直方图 (Histogram)
axes[1, 1].hist(y1, bins=20, color="blue", alpha=0.7, label="Sine Function")  # 绘制正弦函数值分布直方图
axes[1, 1].hist(y_linear, bins=20, color="red", alpha=0.7, label="Linear Function")  # 绘制线性函数值分布直方图
axes[1, 1].set_title("Histogram", fontsize=14)  # 设置标题
axes[1, 1].set_xlabel("y", fontsize=12)  # 设置 x 轴标签
axes[1, 1].set_ylabel("Frequency", fontsize=12)  # 设置 y 轴标签
axes[1, 1].legend(fontsize=10)  # 添加图例
axes[1, 1].grid(True, linestyle="--", alpha=0.7)  # 添加网格线

# 调整子图布局
plt.tight_layout()

# 显示图形
plt.show()
```
